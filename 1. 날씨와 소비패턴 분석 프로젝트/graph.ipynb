```python
app = dash.Dash()
app.layout = html.Div(
    [
    dcc.Dropdown(id = 'inde_val1',
                options = [{'label':i,'value':i}for i in result.columns[6:-1]]),
    html.Br(),
    dcc.Dropdown(id = 'inde_val2',
                options = [{'label':i,'value':i}for i in result.columns[6:-1]]),
    html.Button('그래프',id = 'graph_button',
                n_clicks = 0
    ),
    html.H1(),
    dcc.Graph(id = 'inde_val_graph',
              figure = fig),
    html.H1(),
    dcc.Graph(id = 'heatmap',
             figure = figu),
    html.Br(),
    html.H1(),
    dcc.Dropdown(id = 'sex',
                options = [{'label':i,'value':i}for i in result.sex.unique()]),
    html.Br(),
    dcc.Dropdown(id = 'age',
                options = [{'label':i,'value':i}for i in result.age.unique()]),    
    dcc.Graph(id = 'bar',
             figure = px.line_3d(bar,x='date',y='quant',z = 'sm_pr'))
    ]
)
@app.callback(
    Output('inde_val_graph','figure'),
    Input('graph_button','n_clicks'),
    State('inde_val1','value'),
    State('inde_val2','value'),
)

def update_inde_val_garph(n,inde_val1, inde_val2):
    fig = px.scatter(result,x=inde_val1,y=inde_val2)
    return fig


app.run_server(host = '192.168.0.65',port = '80')
```
