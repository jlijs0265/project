```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score 
result = pd.read_csv('C://pythonworks/miniminiproject/data/buy_2018_base.csv',encoding = 'euc-kr')

# M_20 ~ F_60 까지 나눠야함.
M_20 = result.loc[(result.sex=='M')&(result.age==20)]
# M_20 = result

M_20_a = pd.DataFrame(np.zeros(25).reshape(1,-1))
M_20_a.columns = result.columns
# #M_20~F_60합치기전 빈데이터 하나 만들기

# #고기/생선/달걀/콩류의 것들 합치기
for i in a:
    b = M_20.loc[M_20.sm_pr==i]
    M_20_a=M_20_a.append(b)

M_20_a=M_20_a.iloc[1:,1:-1]

y = M_20_a.groupby('date').sum()['quant']

X = M_20_a.groupby('date').mean()
# y = M_20_a.iloc[:,4:5]
# X = M_20_a.iloc[:,5:]

# Z = X.iloc[:,[0,1,2]]
X = X[['평균기온','강수계속시간', '일강수량', '평균풍속','합계일조시간', '평균중하층운량', 'PM10']]
# Z['date'] = pd.to_datetime(Z['date'])
# Z['date']= Z['date'].astype('int64')

# X['date'] = Z['date']
# X['gender'] = np.where(Z.sex=='M',1,0)


result = [0]
lr_clf = RandomForestClassifier()

for i in np.arange(100):
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size =0.2)
    lr_clf.fit(X_train,y_train)
    acc = accuracy_score(lr_clf.predict(X_test),y_test)
    lr_clf = RandomForestClassifier()
    
    result.append(acc)
print(a,max(result))
```
