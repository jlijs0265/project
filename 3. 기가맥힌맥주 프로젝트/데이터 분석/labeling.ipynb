```python
Lager = list(['Altbier','Bock','Kölsch','Lager','Pilsner'])
Ale = list(['Barleywine','Bitter','Brown Ale','California Common / Steam Beer','Cream Ale','Dubbel',
             'Farmhouse Ale','Gruit / Ancient Herbed Ale','IPA','Mild Ale','Old Ale','Pale Ale',
             'Porter','Pumpkin Beer','Red Ale','Scotch Ale / Wee Heavy','Scottish Ale',
             'Sour','Stout','Strong Ale','Tripel','Wheat Beer','Winter Warmer'])
Sour = list(['Lambic','Sour','Wild Ale'])
```

```python
list(bier.Style.unique())
```

```python
Name = []
Category = []
Class = []
Subclass = []

for i in range(len(bier)):
    if re.split(' - ',list(bier.Style)[i])[0] in Lager:
        Category.append('Lager')
    elif re.split(' - ',list(bier.Style)[i])[0] in Ale:
        Category.append('Ale')
    elif re.split(' - ',list(bier.Style)[i])[0] in Sour:
        Category.append('Sour')
    else: Category.append('Others')
        #Category.append(re.split(' - ',list(bier.Style)[i])[0])
    Name.append(bier.Name[i])
    Class.append(re.split(' - ',list(bier.Style)[i])[0])
    if len(re.split(' - ',list(bier.Style)[i])) == 2:
           Subclass.append(re.split(' - ',list(bier.Style)[i])[1])
    else:
        Subclass.append(re.split(' - ',list(bier.Style)[i])[0])
```

```python
style = pd.DataFrame([Category,Class,Subclass], index=(['Category','Class','Subclass'])).T
```

```python
style.loc[2357]
```

```python
beer = pd.concat([bier,style],axis=1)
```

```python
beer.columns
```

```python
beer = beer[['Name', 'key', 'Style', 'Category', 'Class', 'Subclass', 'Style Key', 'Brewery', 'Description', 'ABV',
       'Ave Rating', 'Min IBU', 'Max IBU', 'Astringency', 'Body', 'Alcohol',
       'Bitter', 'Sweet', 'Sour', 'Salty', 'Fruits', 'Hoppy', 'Spices',
       'Malty', 'label']]
```
